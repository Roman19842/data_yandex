### Cредняя выручка пользователей тарифов «Ультра» и «Смарт» различаются;

Для начала подсчитаем среднюю выручку. Данные о выручке хрянтся в smart_total и ultra_total. 
будем считатать среднее значение тем же способо что и считали средние минуты , сообщения , интернет.
Посмотрим для начала как распределяется выручка.

```python
smart_total['total'].plot(style='o',color='red',alpha=0.5)
ultra_total['total'].plot(style='o',color='blue',alpha=0.5)
```

Тариф Ультра почти всегда чаще укладывается в сумму абонеской платы(1950), а вот смарт явно чаще переплачивает.Есть даже выбросы до 4500р

Проверим гипотезу средняя выручка различается.

**Нашей нулевой гипотезой будет являться** Н0-средняя выручка тарифа смарт и ультра равны.
**Альтренативная**H1- средняя выручка тарифов отличаются.

Это двухстороняя гипотеза.

Для этого исполузуем метод ttest_ind из библиотеки scipy.
Подсчитаем p-value - это вероятность того , что среднее одной выборки может совпвсть с со средним другой выборки.

```python

ultra_total['total'].median()
```

```python
ultra_total['total'].mean()
```

```python
smart_total['total'].mean()
```

```python
smart_total['total'].median()
```

```python
results = st.ttest_ind(
    smart_total['total'], 
    ultra_total['total'],)

print('p-значение: ', results.pvalue)
alpha = 0.05
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Принимаем нулевую гипотезу')
```

Значение p-значение говорит о том ,что среднее значение одной выборки с большой долей вероятности лежит не в 5%  (alpha значение 0.05)диапазоне попаданя к среднему другой выборки.Поэтому гипотезу отвергаем

**Вывод по гипотезе. С большой долей верояности что среднеи обоих выборок не совпадают(не совпадут). Оно кстати видно хорошо визульно**.
